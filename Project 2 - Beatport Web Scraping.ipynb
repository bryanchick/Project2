```python
import requests
website_url = requests.get('https://www.beatport.com/genre/progressive-house/15/top-100').text

from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url, 'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('div',{'class':'bucket tracks top-hundred-tracks'})
My_table
```

```python
#links2 = My_table.findAll('p',{'class':'buk-track-artists'})
#links2
artists = []
for artist in My_table.find_all('a', href=True):
    #print (artist['href'])
    artists.append(artist.get('href'))
print(artists)
```
